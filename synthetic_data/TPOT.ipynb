```python
# Import required libraries
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
import pandas as pd 
import numpy as np
```

```python
# Load the data
data = pd.read_csv('../raw_data/brain_proteomics_data.csv')
data.head(5)
```

```python
data.outcome.value_counts()
```

```python
data.rename(columns={'outcome': 'class'}, inplace=True)
```

```python
data.dtypes
```

```python
# Create X and y
# X is only the columns of proteins, 'years_to_birth','gender', remain

data['gender'] = data['gender'].map({'male':0,'female':1})

X = data.drop(columns =['Case','histological_type','race','ethnicity','radiation_therapy','Grade','Mutation.Count','Percent.aneuploidy','IDH.status','class'])
y = data['class']

```

```python
data['gender'] = data['gender'].map({'male':0,'female':1})

X = data.drop(columns =['Case','years_to_birth','gender','histological_type','race','ethnicity','radiation_therapy','Grade','Mutation.Count','Percent.aneuploidy','IDH.status','class'])
y = data['class']
```

```python
# Split train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
tpot = TPOTClassifier(verbosity=2, generations=5, max_time_mins=10, population_size=30)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))
tpot.export('tpot_pipeline.py')
```
